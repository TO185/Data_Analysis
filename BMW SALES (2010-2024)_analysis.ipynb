```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

**PROJECT 2 -BMW SALES (2010-2024)**

```python
df=pd.read_csv('/content/BMW sales data (2010-2024) (1).csv')
df
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.info()
```

**1) Which car model has the highest sales volume?**

```python
df.groupby('Model')['Sales_Volume'].count()

a=sns.barplot(data=df,x='Model',y='Sales_Volume',hue='Model',palette='viridis',errorbar=None)

for i in a.containers:
  a.bar_label(i,fmt='%.0f')

plt.title('Sales Volume by Model')
plt.xlabel('Model')
plt.ylabel('Sales Volume')
plt.tight_layout()
plt.show()
```

The **BMW X1**  is the top-selling model with the highest total sales volume across all regions. It consistently ranks high due to its blend of performance, luxury, and affordability.

**2) Which region contributes the most to total BMW sales?**

```python
df.groupby('Region')['Sales_Volume'].sum()

a=sns.barplot(data=df,x='Region',y='Sales_Volume',hue='Region',palette='viridis',errorbar=None)

for i in a.containers:
  a.bar_label(i,fmt='%.1f')

plt.title('Total Sales Volume by Region')
plt.xlabel('Region')
plt.xticks(rotation=90)
plt.ylabel('Total Sales Volume')
plt.tight_layout()
plt.show()
```

**North America** is most sales by Region and contributing BMW sales

**3) Which fuel type is most in demand?**

```python
fuel_type=df.groupby('Fuel_Type')['Sales_Volume'].sum()

explode = [0.1 if i == fuel_type.idxmax() else 0 for i in fuel_type.index]

plt.pie(fuel_type,labels=fuel_type.index,autopct='%1.1f%%',explode=explode)
plt.title('Most In Demand Fuel Type')
plt.show()
```

**Hybrid** remains the most in-demand fuel type

**4) Which transmission type is preferred by buyers?**

```python
df.groupby('Transmission')['Sales_Volume'].sum()

a=sns.barplot(data=df,x='Transmission',y='Sales_Volume',hue='Transmission',palette='viridis',errorbar=None)

for i in a.containers:
  a.bar_label(i,fmt='%.1f')

plt.title('Preferred Transmission Type')
plt.xlabel('Transmission')
plt.ylabel('Sales Volume')
plt.title('Preferred Transmission Type')
plt.show()
```

**Automatic transmission** is overwhelmingly preferred

**5) Which car model has the highest average price?**

```python
df.groupby('Model')['Price_USD'].mean()

a=sns.barplot(data=df,x='Model',y='Price_USD',hue='Model',palette='viridis',errorbar=None)

for i in a.containers:
  a.bar_label(i,fmt='%.0f')

plt.title('Average Price by Model')
plt.xlabel('Model')
plt.ylabel('Average Price')
plt.tight_layout()
plt.show()
```

The **BMW 3 Series** have the highest average price

**6) Which sales classification category has the most records, and what are the characteristics of those cars?**

```python
df.groupby('Sales_Classification')['Sales_Volume'].sum()

a=sns.barplot(data=df,x='Sales_Classification',y='Sales_Volume',hue='Sales_Classification',errorbar=None,palette='viridis')

for i in a.containers:
  a.bar_label(i,fmt='%.0f')

plt.title('Total Sales Volume by Sales Classification')
plt.ylabel('Sales Volume')
plt.xlabel('Sales Classification')
plt.show()
```

The **High** Sales Classification category has the most records
